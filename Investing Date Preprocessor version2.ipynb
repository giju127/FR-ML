```python
import pandas as pd
import numpy as np
import xlwings as xw
from datetime import datetime
```

```python
def excel(name):
    app = xw.App(visible=False)
    wb = xw.Book(name)
    sheet = wb.sheets[0]
    df = sheet.range('A1').options(pd.DataFrame, expand='table').value
    app.kill()
    print('Excel load success')
    return df
```

```python
loc = "C:\\Users\\Erion\\Desktop\\Financial Rainmakers\\33ê¸° ML\\Data\\Pure stock market index\\"
name = "MSCI Emerging Markets Data 2012-01-30"
```

```python
data = pd.read_csv(loc+name+".csv")
```

```python
data['Date'] = data['Date'].apply(lambda x: x.replace(',',''))
data['Date'] = data['Date'].apply(lambda x: pd.to_datetime(x, format='%b %d %Y'))
```

```python
data.index = data["Date"]
data = data.drop(["Date"], axis='columns')
data = data.sort_index(ascending=True)
```
